The purpose of this notebook is to visualize features associated with COVID-19 that may correlate to higher case rates including:

    Population over 65
    Adult Obesity
    Adult Smokers
    Adult Diabetes
    Life Expectancy
    Vaccinations

```python
import numpy as np
import matplotlib.pylab as plt
from pandas import read_csv
import matplotlib.pyplot
from pandas import DataFrame
from pandas import Grouper
from pandas import merge
import pandas as pd
```

```python
data = pd.read_csv("popden.csv")
```

```python
data = data[pd.to_datetime(data['date'])=="2020-10-31"]
```

```python
for col in data.columns:
    print(col)
```

```python
data['Cases per 100,000']=(data['cases']/data['total_population'])*100000
```

```python
data = data[(data['cases']>1000)]
data = data[(data['cases']<=10000)]
```

```python
x = data['Cases per 100,000']
y = data['percent_smokers']
plt.plt.rcParams["figure.figsize"] = (20,10)
plt.scatter(x,y)
plt.title('COVID Cases per 100,000 vs. Percent Smokers')
plt.show()
```

```python
x = data['Cases per 100,000']
y = data['percent_uninsured']
plt.plt.rcParams["figure.figsize"] = (20,10)
plt.scatter(x,y)
plt.title('COVID Cases per 100,000 vs. Percent Uninsured')
plt.show()
```
